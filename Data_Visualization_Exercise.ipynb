```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pizza_df = pd.read_excel('/content/PizzaWeek.xlsx')
pizza_df.head()
```

```python
pizza_df.info()
```

```python
plt.style.use('seaborn-dark')
colors = ['Grey','Blue','Orange']
pizza_df.plot(x="Topping", y=['Total','Male','Female'], kind="bar", figsize = (15,12), color = colors)
plt.grid(axis = 'y', c = 'grey', linewidth = 1)
plt.xlabel('Toppings', fontsize = 15)
plt.ylabel('Percentage', fontsize = 15)
plt.title('What Toppings Do You Like On Your Pizza?', fontsize = 20)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.ylim(0, 70)
plt.legend(fontsize = 12)

plt.savefig('PizzaToppings.png', dpi = 300);
```
